```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# CPF
Based on: https://www.cpf.gov.sg/Members/AboutUs/about-us-info/cpf-interest-rates

Simple projections:
1. Lets only consider the first 60k that you gain an additional 1% p.a.
    - Ordinary Account (OA) 20k with 3.5% p.a.
    - Special Account (SA) 40k with 5% p.a.
    - Ignore the re-invested extra amount
2. Assume that the interest rates is constant
    - Based on their historical interest rates, I believe that this is reasonable to assume
    - https://www.cpf.gov.sg/askjamie/user/uploads/Historical%20interest%20rates%20%202019%20Q4.pdf
3. Interest rates are paid annually
    - https://www.cpf.gov.sg/members/FAQ/schemes/other-matters/others/FAQDetails?category=other+matters&group=Others&ajfaqid=2192131&folderid=13726

# Special Account only @ 40k
Since you can only withdraw your Speical Account at 55, I will draw the projection until the year I turn 55 (Current age: 26).

```python
sa_interest = 0.05 # Special Account interest + 1
time_steps = np.arange(0, 55-26+1) # total number of years (time step)
multiplier = (sa_interest * time_steps)+1 # balance as a ratio
sa_balance = multiplier * 40000 # Total amount you will have after each time step

print(f'Now, I will have: ${np.round(sa_balance[0], 2)}')
print(f'In 10 years, I will have: ${np.round(sa_balance[10], 2)}')
print(f'At 20 years, I will have: ${np.round(sa_balance[20], 2)}')
print(f'At 55, I will have: ${np.round(sa_balance[-1], 2)}')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=time_steps, y=sa_balance)
plt.title('Projected Savings considering only 40k in my Special Account over 29 years')
plt.ylabel('Balance ($)')
plt.xlabel('Years')
plt.show()
```

# Wrap this into a function

```python
def calculate_balance(rate, steps, initial_amt):
    interest = rate # Special Account interest + 1
    time_steps = np.arange(0, steps+1) # total number of years (time step)
    multiplier = (interest * time_steps)+1 # compounding the sa_interest
    balance = multiplier * initial_amt # Total amount you will have after each time step
    
    print(f'Now, I will have: ${np.round(balance[0], 2)}')
    print(f'In 10 years, I will have: ${np.round(balance[10], 2)}')
    print(f'At 20 years, I will have: ${np.round(balance[20], 2)}')
    print(f'At 55, I will have: ${np.round(balance[-1], 2)}')
    
    return balance
```

# Ordinary Account @ 20k

```python
steps = 29 # 29 years
oa_balance = calculate_balance(0.035, steps, 20000)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(steps+1), y=oa_balance)
plt.title('Projected Savings considering only 20k in my Ordinary Account over 29 years')
plt.ylabel('Balance ($)')
plt.xlabel('Years')
plt.show()
```

# OA + SA accounts @ 60k

```python
combined_balance = sa_balance + oa_balance
print(f'Now, I will have: ${np.round(combined_balance[0], 2)}')
print(f'In 10 years, I will have: ${np.round(combined_balance[10], 2)}')
print(f'At 20 years, I will have: ${np.round(combined_balance[20], 2)}')
print(f'At 55, I will have: ${np.round(combined_balance[-1], 2)}')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(steps+1), y=combined_balance)
plt.title('Projected Savings considering only 60k in both Ordinary and Special Accounts over 29 years')
plt.ylabel('Balance ($)')
plt.xlabel('Years')
plt.show()
```

# Consolidating our findings
1. In 29 years, you can have 138k if you max out the bonus interest in your CPF Ordinary and Special Account. That is 2.3x your initial investment of 60k. 
2. Close to risk-free.
3. Excludes re-investing the amounts you earn each year from the rates.
4. Assume constant interest rates which might not always be the case.

# Re-investing
Lets edit the function to accept reinvesting

```python
def calculate_balance(initial_amt, steps, rate_1, reinvest=False, limit=None, rate_2=None):
    # If there isn't a 2nd rate, we reuse the 1st rate
    if rate_2 == None:
        rate_2 = rate_1
    
    balance = [initial_amt] # amount at each time step
    current_balance = initial_amt # current time step balance
    
    # calculate maximum interest rate to earn from rate 1
    if reinvest:
        max_top_ups = limit * rate_1
    else: 
        max_top_ups = initial_amt * rate_1
    
    # looping through the time steps
    for s in range(1, steps+1):
        if reinvest:
            # checking limit to multiply correct interest rates
            if current_balance <= limit:
                current_balance = current_balance * (rate_1+1) # calculating the reinvested rates
            else:
                extra_payouts = (current_balance-limit) * rate_2 # calculating the reinvested rates
                # adding the constant top-ups and the reinvested rates
                current_balance = current_balance + max_top_ups + extra_payouts
        else:
            current_balance += max_top_ups # fixed top-ups
        balance.append(current_balance)
    
    print(f'Now, I will have: ${np.round(balance[0], 2)}')
    print(f'In 10 years, I will have: ${np.round(balance[10], 2)}')
    print(f'At 20 years, I will have: ${np.round(balance[20], 2)}')
    print(f'At 55, I will have: ${np.round(balance[-1], 2)}')
    
    return np.array(balance)
```

```python
# special account balance with reinvesting
sa_balance = calculate_balance(initial_amt=40000, steps=29, rate_1=0.05, 
                  reinvest=True, limit=40000, rate_2=0.04)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(steps+1), y=sa_balance)
plt.title('Projected Savings considering only 40k in my Special Account over 29 years while reinvesting interest payouts')
plt.ylabel('Balance ($)')
plt.xlabel('Years')
plt.show()
```

```python
# ordinary account balance with reinvesting
oa_balance = calculate_balance(initial_amt=20000, steps=29, rate_1=0.035, 
                  reinvest=True, limit=20000, rate_2=0.025)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(steps+1), y=oa_balance)
plt.title('Projected Savings considering only 20k in my Ordinary Account over 29 years while reinvesting interest payouts')
plt.ylabel('Balance ($)')
plt.xlabel('Years')
plt.show()
```

```python
combined_balance = sa_balance + oa_balance
print(f'Now, I will have: ${np.round(combined_balance[0], 2)}')
print(f'In 10 years, I will have: ${np.round(combined_balance[10], 2)}')
print(f'At 20 years, I will have: ${np.round(combined_balance[20], 2)}')
print(f'At 55, I will have: ${np.round(combined_balance[-1], 2)}')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(steps+1), y=combined_balance)
plt.title('Projected Savings considering only 60k in both Ordinary and Special Accounts over 29 years while reinvesting interest payouts')
plt.ylabel('Balance ($)')
plt.xlabel('Years')
plt.show()
```

# Consolidating our findings
1. In 29 years, you can have 195k if you max out the bonus interest in your CPF Ordinary and Special Account with reinvested interest payouts. That is 3.25x your initial value.
2. Close to risk-free.
3. Compounded interest brings exponential growth in value
4. Assume constant interest rates which might not always be the case.

# What happens if you keep till 65?

```python
steps = 39 # 39 years
# special account balance with reinvesting
sa_balance = calculate_balance(initial_amt=40000, steps=39, rate_1=0.05, 
                  reinvest=True, limit=40000, rate_2=0.04)
```

```python
# ordinary account balance with reinvesting
oa_balance = calculate_balance(initial_amt=20000, steps=39, rate_1=0.035, 
                  reinvest=True, limit=20000, rate_2=0.025)
```

```python
combined_balance = sa_balance + oa_balance
print(f'Now, I will have: ${np.round(combined_balance[0], 2)}')
print(f'In 10 years, I will have: ${np.round(combined_balance[10], 2)}')
print(f'At 20 years, I will have: ${np.round(combined_balance[20], 2)}')
print(f'At 30 years, I will have: ${np.round(combined_balance[30], 2)}')
print(f'At 55, I will have: ${np.round(combined_balance[-1], 2)}')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(steps+1), y=combined_balance)
plt.title('Projected Savings considering only 60k in both Ordinary and Special Accounts over 29 years while reinvesting interest payouts')
plt.ylabel('Balance ($)')
plt.xlabel('Years')
plt.show()
```

# Consolidating our findings
1. As the reinvested interest brings exponential growth, the amount you earn from the interest widens every subsequent year. You earned almost 80k in the final 10 years.
